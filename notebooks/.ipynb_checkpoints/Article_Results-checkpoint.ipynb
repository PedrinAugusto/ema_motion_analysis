```python
# Plot Angle
%matplotlib inline
plt.figure()
plt.plot(time_vec_cycle_target, angles_vec_knee_target, label='Knee')
plt.plot(time_vec_cycle_target, angles_vec_elb_target, label='Elbow')
plt.ylabel("Angle (degrees)")
plt.xlabel("Time (s)")
plt.legend()
plt.grid(True)
plt.show()
```

## Segmenting Cycles Using the Knee Angle

```python
peaks, properties = find_peaks(-angles_vec_knee_target, height=-60)
plt.plot(time_vec_cycle_target, angles_vec_knee_target)
plt.vlines(x=time_vec_cycle_target[peaks], ymin=min(angles_vec_knee_target), 
           ymax = max(angles_vec_knee_target), color = "black", linestyles ='dashed')
plt.ylabel("Angle (degrees)")
plt.xlabel("Time (s)")
plt.grid(True)
plt.show()
```

```python
row_cycles = []
for i in range(1, len(peaks)):
    row_cycles.append(np.array(angles_vec_target_sep[peaks[i-1]-1:peaks[i],:]))
```

```python
%matplotlib inline
def f(n):
    plt.figure()
    for i in range(len(angles_names)):
        plt.plot(row_cycles[n][:,i], label=angles_names[i])
    plt.legend()
    plt.grid(True)
    plt.ylabel("Angle (degrees)")
    plt.xlabel("Frame [n]")
    plt.show()
    
cycle = wg.IntSlider(description='cycle number', max=len(row_cycles)-1, continuous_update=False)
out = wg.interactive_output(f, {'n': cycle})

display(cycle, out)
```

## Applying DTW to every cycle

```python
estimated_cycles = []
distances_knee = []
distances_elb = []
for n in range(len(row_cycles)):
    distance_knee, paths_knee = fastdtw(angles_vec_knee_ref, row_cycles[n][:,angles_names.index(angle_knee)], dist=euclidean)
    distance_elb, paths_elb = fastdtw(angles_vec_elb_ref, row_cycles[n][:,angles_names.index(angle_elb)], dist=euclidean)
    paths_knee = np.array(paths_knee)
    paths_elb = np.array(paths_elb)
    estimated_targets = [0, 0, 0, 0, len(row_cycles[n][:,angles_names.index(angle_knee)])-1]
    i=1
    idx_path = np.mean(np.where(paths_elb[:, 0] == pose_frame_ref[i] - pose_frame_ref[0])[0]).astype(int)
    idx_target = paths_knee[idx_path, 1]
    estimated_targets[i] = idx_target
    for i in (2,3):
        idx_path = np.mean(np.where(paths_elb[:, 0] == pose_frame_ref[i] - pose_frame_ref[0])[0]).astype(int)
        idx_target = paths_elb[idx_path, 1]
        estimated_targets[i] = idx_target
    estimated_cycles.append(estimated_targets)
    distances_knee.append(distance_knee)
    distances_elb.append(distance_elb)
```

```python
def interactiveRowingMoments(n):
    plotRowingMoments(row_cycles[n], estimated_cycles[n], fps_target, angles_names)
    
cycle = wg.IntSlider(description='cycle number', max=len(row_cycles)-1, continuous_update=False)
out = wg.interactive_output(interactiveRowingMoments, {'n': cycle})
display(cycle, out)
```

## Interactive Interface for Training

```python
rowing_phases = ['Catch', 'Leg Drive', 'Arm Drive', 'Arm Recovery', 'Leg Recovery']
def showRowingPhaseTraining(row_phase, cycle_n, frame_n):  
    frame_n_abs = frame_n + n0
    frame, frame_width, frame_height = getFrame(video_path_target, frame_n_abs, input_path=True)
    angles = angles_vec_target[frame_n_abs]
    pose_keypoints = keypoints_vec_target[frame_n_abs]
    image = poseDATAtoFrame(frame, pose_keypoints, 0, joint_names, pairs_names, color=0)
    angles_image = showAnglesRowing(image, pose_keypoints, angles, thickness=1, textsize=0.5, adj=5, alpha=1.0)
    
    tab_len = round(frame_width*0.5)
    title_color = (255,255,255)
    info_color = (50,170,50)
    text_thickness = 1
    text_size = 0.8
    h_space = 20
    h_space_init = 40
    v_space = 38
    v_space_init = round(v_space/1.2)
    figsize=[23,15]
    
    data_image = np.zeros([angles_image.shape[0], angles_image.shape[1] + tab_len, angles_image.shape[2]])
    data_image[:, :angles_image.shape[1], :] = angles_image[:,:,:]

    cv2.putText(data_image, "Cycle Analysis".format(frame_n*(1/fps_target)), 
                (angles_image.shape[1]+h_space_init,v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size*1.25, title_color, text_thickness)
    cv2.putText(data_image, "Video Time: {0:.2f} s".format(frame_n*(1/fps_target)), 
                (angles_image.shape[1]+h_space,1*v_space+v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size, info_color, text_thickness)
    cv2.putText(data_image, "Cycle Time: {0:.2f} s".format((frame_n-peaks[cycle_n])*(1/fps_target)), 
                (angles_image.shape[1]+h_space,2*v_space+v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size, info_color, text_thickness)
    cycle_period = (estimated_cycles[cycle_n][4]-estimated_cycles[cycle_n][0])*(1/fps_target)
    cv2.putText(data_image, "Period: {0:.2f} s".format(cycle_period), 
                (angles_image.shape[1]+h_space,3*v_space+v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size, info_color, text_thickness)
    pace_spm = round(60/cycle_period)
    cv2.putText(data_image, "Pace: {} s/m".format(pace_spm), 
                (angles_image.shape[1]+h_space,4*v_space+v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size, info_color, text_thickness)
    drive_time = (estimated_cycles[cycle_n][2]-estimated_cycles[cycle_n][0])*(1/fps_target)
    drive_pct = round(100*drive_time/cycle_period)
    cv2.putText(data_image, "Drive: {} %".format(drive_pct), 
                (angles_image.shape[1]+h_space,5*v_space+v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size, info_color, text_thickness)
    recovery_pct = 100-drive_pct
    cv2.putText(data_image, "Recovery: {} %".format(recovery_pct), 
                (angles_image.shape[1]+h_space,6*v_space+v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size, info_color, text_thickness)
    for k in range(len(angles_names)):
        angle = angles_names[k]
        correct_angle = angles_vec_ref[pose_frame_ref[rowing_phases.index(row_phase)], angles_names.index(angle)]
        cv2.putText(data_image, "Error {}: {}".format(angle.split(" ")[2], int(round(angles[k]) - correct_angle)), 
                (angles_image.shape[1]+h_space,(7+k)*v_space+v_space_init), cv2.FONT_HERSHEY_COMPLEX, text_size, info_color, text_thickness)
    
    showFrame(data_image.astype(int), figsize=figsize)

phase_dp = wg.Dropdown(options=rowing_phases,
                        value='Catch',
                        description='Phase: ',
                        disabled=False)
cycle_slider = wg.IntSlider(
    value=0,
    min=0,
    max=len(row_cycles)-1,
    step=1,
    description='Cycle: ',
    continuous_update=False,
)
frame_slider = wg.IntSlider(value=0,
        min=0,
        max=n_frames_target-1,
        step=1,
        description='Frame:',
        disabled=False,
        continuous_update=False,
        orientation='horizontal',
        readout=True,
        readout_format='d'
    )
frame_n = wg.Play(
        value=estimated_cycles[cycle_slider.value][rowing_phases.index(phase_dp.value)] + peaks[cycle_slider.value],
        min=0,
        max=target_len-1,
        step=1,
        description="Press play",
        disabled=False
    )
wg.jslink((frame_n, 'value'), (frame_slider, 'value'))

def onPhaseChange(change):
    frame_n.value = estimated_cycles[cycle_slider.value][rowing_phases.index(phase_dp.value)] + peaks[cycle_slider.value]
def onCycleChange(change):
    frame_n.value = estimated_cycles[cycle_slider.value][rowing_phases.index(phase_dp.value)] + peaks[cycle_slider.value]

phase_dp.observe(onPhaseChange, names='value')    
cycle_slider.observe(onCycleChange, names='value')

out_rp = wg.interactive_output(showRowingPhaseTraining, {'row_phase': phase_dp,
                                                         'cycle_n': cycle_slider,
                                                         'frame_n': frame_n})
hbox = wg.HBox([cycle_slider, phase_dp])
hbox_player = wg.HBox([frame_n, frame_slider])
display(hbox, out_rp, hbox_player)
```

# Phases Analysis

```python
# Hip angle during each phase
%matplotlib inline
def phaseAnalysis(phase, angle):
    plt.figure()
    phase_idx = np.array(estimated_cycles)[:, rowing_phases.index(phase)] + peaks[:-1]
    plt.plot(angles_vec_target_sep[phase_idx, angles_names.index(angle)], 
             label = angle.split(" ")[2] + " Angle Target")
    plt.plot(np.ones(len(estimated_cycles))*angles_vec_ref[pose_frame_ref[rowing_phases.index(phase)], angles_names.index(angle)], 
             label = angle.split(" ")[2] + " Angle Reference", linestyle='dashed')
    plt.legend()
    plt.grid(True)
    plt.ylabel("Angle (degrees)")
    plt.xlabel("Cycle [n]")
    plt.show()
    
    

phase_dp = wg.Dropdown(options=rowing_phases,
                        value='Catch',
                        description='Phase: ',
                        disabled=False)
angle_dp = wg.Dropdown(options=angles_names,
                        value=angles_names[1],
                        description='Phase: ',
                        disabled=False)

hbox = wg.HBox([phase_dp, angle_dp])
out = wg.interactive_output(phaseAnalysis, {'phase': phase_dp, 'angle': angle_dp})

display(hbox, out)
```
